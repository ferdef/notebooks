# Active Organizations

```sql
WITH MaxDates(iid, maxdate) AS (
    SELECT
        po.iid,
        MAX(pca.dusagedate) as maxdate
    FROM dd_customers c
    JOIN dd_plasticorder po ON po.fidcustomer = c.iid
    JOIN dd_plasticcloud_organization pco ON pco.fidorder = po.iid
    LEFT JOIN dd_plasticcloudactivity pca ON pca.fidorder = po.iid
    WHERE 
        po.bactive = 'T'
    GROUP BY po.iid
) 

SELECT 
    maxdate,
    Count(*)
FROM MaxDates
GROUP BY maxdate
ORDER BY maxdate DESC
```

# Active Organizations by Region

```sql
WITH MaxDates(iid, region, maxdate) AS (
    SELECT
        po.iid,
        pco.fidregion,
        MAX(pca.dusagedate) as maxdate
    FROM dd_customers c
    JOIN dd_plasticorder po ON po.fidcustomer = c.iid
    JOIN dd_plasticcloud_organization pco ON pco.fidorder = po.iid
    LEFT JOIN dd_plasticcloudactivity pca ON pca.fidorder = po.iid
    WHERE 
        po.bactive = 'T'
    GROUP BY
        pco.fidregion,
        po.iid
)

SELECT
    Max(MaxDates.maxdate) as maxdate,
    region
FROM MaxDates
GROUP BY MaxDates.region
ORDER BY maxdate DESC
```
